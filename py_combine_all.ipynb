```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning) 

cols = ["L", "t_gsl_s", "t_g4_s", "diff2"]

data_table = pd.read_csv('./Downloads/output.csv', comment = "#", header = None, names = cols)
data_table
```

```python
L = data_table['L']
gs = data_table['t_gsl_s']
g4 = data_table['t_g4_s']
diff2 = data_table['diff2']
```

```python
fig,ax = plt.subplots(1,1)                  
fig.set_size_inches(11,8.5)                   

fig.tight_layout()                            

x = np.linspace(0, 300, 20)
y = np.linspace(0,500, 20)
plt.yscale('log')

ax.set_xlabel("L")
ax.set_ylabel("Time (s)") 

for value in data_table:
    
    ax.plot(L, gs, marker="o",linestyle="None",markersize=7,label=value)
    ax.plot(L, g4, marker="o",linestyle="None",markersize=7,label=value)
    

ax.legend(("t_gsl_s", "t_g4_s"));
```
