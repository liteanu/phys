```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

```

```python
cols = ["L", "t_gsl_s", "t_g4_s", "diff2"]

data_table = pd.read_csv('./Downloads/output.csv', comment = "#", header = None, names = cols)
data_table
```

```python
L = data_table['L']
gs = data_table['t_gsl_s']
g4 = data_table['t_g4_s']
diff2 = data_table['diff2']
```

```python
fig,ax = plt.subplots(1,1)                    # One window
fig.set_size_inches(11,8.5)                   # (width,height) - letter paper landscape

fig.tight_layout()                            # Make better use of space on plot

ax.set_xlim(0.0,9.0)
ax.set_ylim(0.0,0.22)

ax.set_xlabel("This is L")
ax.set_ylabel("This is t_gsl_s")

for value in data_table:
    
    ax.plot(L, gs, marker="o",linestyle="None",markersize=7,label=value)
    
ax.legend(loc=0,shadow=True);
```
