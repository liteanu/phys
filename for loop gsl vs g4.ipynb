```c++
#include <iostream>
#include "G4LegendrePolynomial.hh"
#include <random>
#include <cstdlib>
#include "gsl/gsl_sf_legendre.h"
#include <stdio.h> 
#include <stdlib.h> 
#include <chrono>

{
    G4LegendrePolynomial lp; 
    gsl_sf_legendre_t norm = GSL_SF_LEGENDRE_NONE;
    std::random_device rd;
    std::mt19937 mt(rd());
    std::uniform_real_distribution<double> xdist(-1, 1);
    
    {
       for (int l=0; l < 3; ++l){
          std::vector<double> result_array(gsl_sf_legendre_array_n(l), 0);
           for (int i=0; i < 3; ++i){
                std::cout << "Value of l : " << l << std::endl ;
                std::cout << "value of i: " <<  i  << std::endl ;
                double x = xdist(mt);
                gsl_sf_legendre_array_e(norm, l, x, -1, &(result_array[0]));
                std::cout << "value of x: " << x << std::endl ;
                   for (int m = -l; m <= l; ++m){
                       std::cout << "Value of m: " << m << std::endl ;
                       //tg4 start
                       auto started_g4 = std::chrono::high_resolution_clock::now();
                       double alpg4 = lp.EvalAssocLegendrePoly(l, m, x); 
                       std::cout << "from alp g4:" << lp.EvalAssocLegendrePoly(l, m, x) << std::endl;
                       //tg4 end
                       auto done_g4 = std::chrono::high_resolution_clock::now();
                       std::cout << "time for g4:" << std::chrono::duration_cast<std::chrono::nanoseconds>(done_g4-started_g4).count() << " nanoseconds"<< std::endl;
                       
                       //tgsl start
                       auto started_gsl = std::chrono::high_resolution_clock::now();
                       double gsl = result_array[gsl_sf_legendre_array_index(l, m)] ;
                       std::cout << "from gsl:" << result_array[gsl_sf_legendre_array_index(l, m)] << std::endl;
                       //tgsl end
                       auto done_gsl = std::chrono::high_resolution_clock::now();
                       std::cout << "time for gsl:" << std::chrono::duration_cast<std::chrono::nanoseconds>(done_gsl-started_gsl).count() << " nanoseconds"<< std::endl;
                       
                      
               }
           }
       }
    }
}
```
