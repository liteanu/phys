```python
def factorial(n):
    if n>1:
        return n*factorial(n-1);
    else:
        return 1
```

```python
def nCr(l, k):
    val=1
    for i in range(k):
        val*=(l-i)
    value=val/factorial(k); 
    return value
```

```python
def PCP(l, m, k):
    u=pow(-1,m) * pow(2,l) *factorial(k)/factorial(k-m)*nCr(l,k)*nCr((l+k-1)/2, l) 
    return u
```

```python
for l in range(10):  
    for m in range(l+1): 
        print("(1-x^2)^"  ,m ,"/2 * (", end="")  
        for k in range(m,l+1):
            if (k>m):
                print("+", end="")
            print(PCP(l, m, k) ,"*x^" ,(k-m), end="") 
        print("),") 
```
