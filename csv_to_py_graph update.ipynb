```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("mplstyle.txt")
```

```python
cols = ["L", "t_gsl_s", "t_g4_s", "avg_diff2", "avg"]

data_table = pd.read_csv('./Downloads/output_2.csv', comment = "#", header = None, names = cols)
data_table
```

```python
L = data_table['L']
gs = data_table['t_gsl_s']
g4 = data_table['t_g4_s']
avg_diff2 = data_table['avg_diff2']
avg= data_table['avg']
```

```python
fig,ax = plt.subplots(1,1)                  
fig.set_size_inches(11,8.5)                                           

ax.set_yscale('log')  


plt.xlabel('L', fontsize=18)
plt.ylabel('t (sec)', fontsize=16)

ax.plot(L, gs, marker='o',linestyle="None",markersize=7,label='gsl')
ax.plot(L, g4, marker="o",linestyle="None",markersize=7,label='g4')
ax.legend(loc=4, ncol=2, frameon=False)

plt.savefig('time.pdf',bbox_inches='tight') 

```

```python
fig,ax = plt.subplots(1,1)               
fig.set_size_inches(11,8.5)                                         

ax.set_yscale('log')  
ax.set_xlabel("L")
ax.set_ylabel("$ \\langle \Delta^2 \\rangle $")    
ax.plot(L, np.sqrt(avg_diff2*1e18/(1e6*(L+1))), marker="o",linestyle="None",markersize=7)    

plt.savefig('diff.pdf',bbox_inches='tight') 
```
